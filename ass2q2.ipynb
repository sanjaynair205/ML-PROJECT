```python
import numpy as np
import matplotlib.pyplot as plt

# Given dataset
X = np.array([12, 23, 34, 45, 56, 67, 78, 89, 123, 134])
Y = np.array([240, 1135, 2568, 4521, 7865, 9236, 11932, 14589, 19856, 23145])

# Reshape X into a column vector and add a bias term (column of ones)
X_matrix = np.vstack((np.ones(len(X)), X)).T  # Shape (10,2) with bias term
Y_matrix = Y.reshape(-1, 1)  # Shape (10,1)

# Compute theta (regression coefficients) using the formula: (X^T X)^(-1) X^T Y
theta = np.linalg.inv(X_matrix.T @ X_matrix) @ X_matrix.T @ Y_matrix

# Extract coefficients
alpha = theta[0, 0]  # Intercept
beta = theta[1, 0]   # Slope
print(f"Intercept (alpha): {alpha:.4f}")
print(f"Slope (beta): {beta:.4f}")

# Predictions
y_pred = X_matrix @ theta  # Compute predicted Y values

# Calculate SSE (Sum of Squared Errors)
SSE = np.sum((Y - y_pred.flatten()) ** 2)
print(f"Sum of Squared Errors (SSE): {SSE:.4f}")

# Calculate SST (Total Sum of Squares)
SST = np.sum((Y - np.mean(Y)) ** 2)

# Calculate R-squared (R^2) value
R2 = 1 - (SSE / SST)
print(f"R-squared (R^2): {R2:.4f}")

# Visualization
plt.scatter(X, Y, color='red', label='Actual Data')
plt.plot(X, y_pred, color='blue', linestyle='-', label='Linear Regression Fit')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid()
plt.show()
```
