```python
import numpy as np

def gradient_descent(X, Y, learning_rate=0.0001, iterations=10000):
    m, b = 0, 0  # Initializing coefficients
    n = len(X)

    for _ in range(iterations):
        Y_pred = m * X + b


        dm = (-2/n) * sum(X * (Y - Y_pred))
        db = (-2/n) * sum(Y - Y_pred)


        m -= learning_rate * dm
        b -= learning_rate * db

    return m, b

# Dataset
X = np.array([12, 23, 34, 45, 56, 67, 78, 89, 123, 134], dtype=float)
Y = np.array([240, 1135, 2568, 4521, 7865, 9236, 11932, 14589, 19856, 23145], dtype=float)

# Compute regression coefficients
m, b = gradient_descent(X, Y)
print(f"Slope (m): {m}, Intercept (b): {b}")
```
